#### Assuming that the model uses Euclidean distance to find the nearest neighbor, what prediction will the model return for each of the following query instances?
|ID |Wave Size (FT)| Web Period (SECS)| Wind Speed (MPH)| Good Surf|
|---|---|---|---|---|
|Q1|8|15|2|?|
|Q2|8|2|18|?|
|Q3|6|12|4|?|

````python
df = pd.read_csv('ques1.csv', sep=r',')
qdf = pd.read_csv('ques1un.csv', sep=r',')
features = ['Wave Size (FT)', 'Web Period (SECS)', 'Wind Speed (MPH)']
df
````

````python
def eucdist(df, row):
    edist = (df - row)**2
    edist = np.sqrt( edist.apply(np.sum, axis=1).to_list() )
    return edist

for k in range(qdf.shape[0]):
    tag = f'eucdist Q{k+1}'
    d = eucdist(df[features], qdf[features].iloc[k])
    df[tag] = d

df
````

````python
colns = ['eucdist Q1','eucdist Q2','eucdist Q3']
prediction = []

for coln in colns:
    mv = np.min(df[coln]) # minimum distance in the column
    idmv = df[coln].to_list().index(mv) # index of min value in the column
    prediction.append(df.loc[idmv,'Good Surf'])

prediction
qdf['Good Surf'] = prediction
qdf 
````

#### [a]

````python
countdf_a = countdf.copy()
qcountdf_a = qcountdf.copy()

# Euclidian distance between each country and Russia
dist = eucdist(countdf_a[features], qcountdf_a[features].iloc[0])
countdf_a['eucdist Rus'] = dist
countdf_a.head()
````

````python
k = 3
near_neighbors = []
sorted_eucdist = countdf_a['eucdist Rus'].sort_values().to_list()

for k in range(k):
    for val in countdf_a['eucdist Rus']:
        if sorted_eucdist[k] == val:
            idx = countdf_a['eucdist Rus'].to_list().index(sorted_eucdist[k])
            near_neighbors.append(countdf_a.loc[idx,'CPI'])
            # print(countdf_a.iloc[[idx]]) # neareast neighbours

prediction = np.mean(near_neighbors)
qcountdf_a['CPI'] = prediction
qcountdf_a
````
