```python
"""
test one: single function
    y' = y - t^2 + 1
"""
def f1(t, y):
    dy = y - t**2 + 1
    
    return dy


def true_f1(t):
    from numpy import exp
    
    yt = (t + 1)**2 - exp(t)/2
    
    return yt

    
"""
test two: sytem of equtions
    y1' = -4y1 + 3y2 + 6
    y2' = -2.4y1 + 1.6y2 + 3.6
"""
def f2(t,y):
    y1 = y[0]
    y2 = y[1]

    dy1 = -4 * y1 + 3 * y2 + 6
    dy2 = -2.4 * y1 + 1.6 * y2 + 3.6
    dy = np.array([dy1, dy2])

    return dy


def true_f2(t):
    from numpy import exp
    
    yt1 = -3.375*exp(-2*t) + 1.875*exp(-0.4*t) + 1.5
    yt2 = -2.25*exp(-2*t) + 2.25*exp(-0.4*t)
    
    return (yt1, yt2)
```

```python
method = pj.euler
```

```python
# implementing test one
a, b, N = 0, 2, 10
ya = np.array([0.5])
t1, y1 = method(f1, a, b, ya, N)
# print(f't1: {t1}')
# print(f'y1: {y1}')

yt = true_f1(t1)

# plotting test one's numerical solution
plt.plot(t1, y1[0], marker='o', label='Pred y', color='navy')

# plotting the true solution to the second function
plt.plot(t1, yt, marker='*', label='True y', color='darkgreen')
plt.legend()
plt.show()
```

```python
# visualizing error
err = np.abs(yt - y1[0]) # |y - y*|
plt.plot(t1, err, marker='o', label='|y - y*|', color='red')
plt.xlabel('t')
plt.ylabel('absolute error')
plt.legend()
plt.show()
```

```python
# implementing test two
a, b, N = 0, 1, 10
ya = np.array([0, 0])
t2, y2 = method(f2, a, b, ya, N)
# print(f't2: {t2}')
# print(f'y2_1: {y2[0,:]}')
# print(f'y2_2: {y2[1,:]}')

yt1, yt2 = true_f2(t2)

# plotting test two's numerical solution
plt.plot(t2, y2[0,:], marker='o', label='Pred y1', color='navy')
plt.plot(t2, y2[1,:], marker='>' , label='Pred y2', color='navy')

# plotting the true solution to the second function
plt.plot(t2, yt1, marker='*', label='True y1', color='darkgreen')
plt.plot(t2, yt2, marker='d' , label='True y2', color='darkgreen')
plt.ylabel('y')
plt.xlabel('t')
plt.legend()
plt.show()
```

```python
# visualizing errors
err1 = np.abs(yt1 - y2[0,:]) # |y1 - y1*|
err2 = np.abs(yt2 - y2[1,:]) # |y2 - y2*|
plt.plot(t2, err1, marker='o', label='|y1 - y1*|', color='red')
plt.plot(t2, err2, marker='>', label='|y2 - y2*|', color='red')
plt.xlabel('t')
plt.ylabel('absolute error')
plt.legend()
plt.show()
```
